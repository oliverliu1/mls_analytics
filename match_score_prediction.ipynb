```python
import pandas as pd
from pathlib import Path
```

```python
path = Path.home() / 'Downloads' / 'mls'
p = path.glob('**/*')
files = [x for x in p if x.is_file()]
```

```python
file_path = path / 'matches.csv'
df_matches = pd.read_csv(file_path)
```

```python
df_matches.info()
```

```python
def get_result(home_score, away_score):
    if home_score > away_score:
        return 'W'
    elif home_score < away_score:
        return 'L'
    else:
        return 'D'
```

```python
df_matches['home_result'] = df_matches.apply(lambda row: get_result(row['home_score'], row['away_score']), axis=1)
```

```python
# Remove matches that went into a shootout
df_matches = df_matches[df_matches.shootout.isnull()]
```
